# EDA
## Introduction
After cleaning the data it is important to examine it to see if there is class imbalance, correlations, separations and if the data falls into any distributions.
## Input
The cleaned data from the data cleaning notebook.
## Output
1. Bar chart showing species distribution
2. Pair plot showing the relationship between the features and species
3. Correlation Matrix
4. Box Plot

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
iris_df = pd.read_csv('data/iris_clean.csv', skipinitialspace = True)
iris_df
```

## The species are distributed evenly.
We don't need to deal with species imbalance.

```python
sns.countplot(x='species', data=iris_df);
```

## Pair Plot
1. There is fair amount of overlap for sepal length values.
2. There is a large amount of overlap for sepal width values.
3. Petal length has distinct values for setosa; there is some overlap for versicolor and virginica.
4. There is some overlap of petal width but the values are more distinct than the other features.
5. There is very good separation of the species for petal length and petal width

```python
sns.pairplot(iris_df, hue = 'species');
```

## Correlation
1. Petal width and petal length are highly correlated
2. Petal width and sepal length are highly correlated
3. Petal length and sepal legnth are highly correlated

```python
corr = iris_df[iris_df.columns[:-1]].corr()
```

```python
sns.set(rc={"figure.figsize":(10, 10)});
sns.heatmap(corr, annot = True);
```

## Variance
There is little variance in petal length and petal width for setosa.

```python
fig, axes = plt.subplots(2, 2);
sns.boxplot(data=iris_df, x='species', y='sepal_length', ax=axes[0,0]);
sns.boxplot(data=iris_df, x='species', y='sepal_width', ax=axes[0,1]);
sns.boxplot(data=iris_df, x='species', y='petal_length', ax=axes[1,0]);
sns.boxplot(data=iris_df, x='species', y='petal_width', ax=axes[1,1]);
```
